```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

otest = pd.read_csv("Freyja_Pumpkins.csv")
otrain = pd.read_csv("Gotem Pumpkins.csv")

test = otest.copy()
train = otrain.copy()

def map_p(type):
    if type == "Çerçevelik":
        return 1
    else: return 0

deets = np.array([list(value) for value in train.drop("Class", axis=1).values]) 
classes = np.array([val for val in map(map_p, list(train['Class']))])

tdeets = np.array([list(value) for value in test.drop("Class", axis=1).values]) 
tclasses = np.array([val for val in map(map_p, list(test['Class']))])

plt.figure(figsize=(10,10))
dataplot = sb.heatmap(otrain.replace({"Çerçevelik": 1, "Ürgüp Sivrisi": 0}).infer_objects(copy=False).corr(), cmap="YlGnBu", annot=True)
```

## Defining the Functions
Sigmoid, z-score normalisation, loss/cost, gradient and gradient descent functions for Logistic Regression

```python
def sigmoid(z):
    g = 1/(1+np.exp(-z)) 
    return g

def znormal(X):
    #Z score normalisation algorithm. Returns z score normalised dataset, means and std. deviations of the features.
    # find the mean of each features columns
    means     = np.mean(X, axis=0)                 # mean will have length features
    # find the standard deviation of each features columns
    stdevs  = np.std(X, axis=0)+0.00001                 # sigma will have length features. +0.00001 to prevent divide by 0 error.
    X_norm = (X - means) / stdevs  
    return (X_norm, means, stdevs)

def compute_cost_logistic(X, y, w, b, lambdaval):
    examples, features  = X.shape
    cost = 0.
    for i in range(examples):                                     
        f_wb_i = sigmoid(np.dot(X[i], w) + b)                                          
        cost +=  -y[i]*np.log(f_wb_i) - (1-y[i])*np.log(1-f_wb_i)   # cost += loss, final cost = summation losses divided by training examples   
    cost = cost/examples                                                      
    
    reg_cost = 0
    for j in range(features):
        reg_cost += (w[j]**2)                                          
    reg_cost = (lambdaval/(2*examples)) * reg_cost             #regularisation                 
    total_cost = cost + reg_cost                                       
    return total_cost           

def compute_gradient_logistic(X, y, w, b, lambdaval): 
    examples, features = X.shape
    jw_derivate = np.zeros((features,))                            
    jb_derivate = 0                                    
    
    for i in range(examples):
        f_wb_i = sigmoid(np.dot(X[i],w) + b)         
        err_i  = f_wb_i  - y[i]                     
        for j in range(features):
            jw_derivate[j] = jw_derivate[j] + err_i * X[i,j]      # summation of errors times example
        jb_derivate = jb_derivate + err_i   #summation of errors
    jw_derivate = jw_derivate/examples                                   
    jb_derivate = jb_derivate/examples                           
    for j in range(features):
        jw_derivate[j] = jw_derivate[j] + (lambdaval/examples) * w[j]
    
    return jb_derivate, jw_derivate

def gradient_descent(X, y, w_in, b_in, alpha, num_iters, lambdaval): 
    J_history = [] #storing cost functions
    nw = w_in.copy()
    b = b_in
    
    for i in range(num_iters):
        jb_derivate, jw_derivate = compute_gradient_logistic(X, y, nw, b, lambdaval)   

        # w = w - alpha * dJ/dw and b = b - alpha * dJ/db
        nw = nw - alpha * jw_derivate               
        b = b - alpha * jb_derivate               

        J_history.append( compute_cost_logistic(X, y, nw, b, lambdaval) )
        if i % (num_iters/20) == 0 or i == num_iters-1:
            print(f"Iteration {i:4d}: Cost {J_history[-1]}   ")
        
    return nw, b, J_history         #return w,b and cost function history for predictions
```

## Getting the weights and biases

```python
w_tmp  = np.zeros_like(deets[0])
b_tmp  = 0.
alph = 3
lambda_ = 0
iters = 15000
w_out, b_out, cost_hist = gradient_descent(znormal(deets)[0], classes, w_tmp, b_tmp, alph, iters, lambda_)
ignore = 0
plt.xlabel('Iterations')
plt.ylabel('Cost Function')
plt.title(f"Cost Function v/s Iterations (first {ignore} values ignored)")
plt.plot(range(ignore, iters), cost_hist[ignore:])
plt.show()
```

## Predicting Outputs

```python
predictions = sigmoid(np.dot(znormal(tdeets)[0], w_out) + b_out)
```

## Accuracy Determination

```python
clamped = [1 if value >= 0.5 else 0 for value in predictions]
TP = 0
TN = 0
FP = 0
FN =  0
for index, predicted in enumerate(clamped):
    got = tclasses[index]
    if predicted > got:
        FP += 1
    elif got > predicted:
        TN += 1
    elif got == predicted:
        if got == 0: FN += 1 
        else: TP += 1

cm = [[TP, FP], [TN, FN]]
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
sb.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Çerçevelik', 'Ürgüp Sivrisi'], yticklabels=['Çerçevelik', 'Ürgüp Sivrisi'])
plt.show()
print(f"The model is {(TP + FN)*100/len(clamped)}% accurate")
```

## Plotting the Sigmoid Graph

```python
pltpredictions = predictions
pltpredictions.sort()
pltclasses = tclasses
pltclasses.sort()
plt.rcParams['figure.figsize'] = [8, 8]
plt.plot(range(1, len(pltpredictions) + 1), pltpredictions)
plt.plot(range(1, len(pltclasses)+1), pltclasses, c = 'r')
plt.plot(range(1, len(tclasses)+1), [0.5 for val in tclasses], c = 'g')
plt.ylabel("Ürgüp Sivrisi                                 Çerçevelik")
plt.xlabel("Training examples (ascending order of predictions)")
plt.grid()
plt.show()
```
